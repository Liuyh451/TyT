```python
import numpy as np

# 加载数据（替换为你的路径）
true_coords = np.load('./dataset/dataset_test.npy')  # shape (425, 2)
# 计算归一化参数
min_vals = true_coords.min(axis=0)
max_vals = true_coords.max(axis=0)

# 对预测值反归一化
pred_denorm = prediction * (max_vals - min_vals) + min_vals
```

```python
import matplotlib.pyplot as plt

def plot_typhoon_track_comparison(true_coords, pred_data):

    lon_pred = pred_data[:, 0]
    lat_pred = pred_data[:, 1]
    # 真实值
    lon_true = true_coords[:, 0]
    lat_true = true_coords[:, 1]

    # 绘图（横轴：纬度，纵轴：经度）
    plt.figure(figsize=(10, 6))
    plt.plot(lat_true, lon_true, marker='o', label='True Track', color='blue')
    plt.plot(lat_pred, lon_pred, marker='x', label='Predicted Track', color='red')

    plt.xlabel("Latitude")
    plt.ylabel("Longitude")
    plt.title("Typhoon Track Comparison")
    plt.xlim(90, 180)
    plt.ylim(0, 65)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
true_data=true_coords[-20:]
pred_data=pred_denorm[-20:]
```

```python
plot_typhoon_track_comparison(true_data, pred_data)
```
